```python
##1.Create a Tensorflow Constant
import tensorflow as tf
values=[1,2,3]
constant_tensor=tf.constant(values)
print(constant_tensor)
```

```python
##2.Perform Element-Wise Operation
import tensorflow as tf
a=tf.constant([2,4,6])
b=tf.constant([1,2,3])
result=tf.multiply(a,b)
print(result)
```

```python
##3.Reshape a Tensor
import tensorflow as tf
x=tf.constant([1,2,3,4,5,6])
result=tf.reshape(x,[2,3])
print(result)
```

```python
##4.Extract a Slice
import tensorflow as tf
x=tf.constant([[1,2,3],[4,5,6],[7,8,9]])
result=(x[:,1])
print(result)
```

```python
##5.Variable Initialization
import tensorflow as tf
weights=tf.constant(tf.random.normal(shape=(3,3)))
print(weights)
```

```python
##6.Math Operations
import tensorflow as tf
x=tf.constant([[25.0,121.0],[625.0,64.0]])
result=tf.sqrt(x)
print(result)
```
